```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from keras.utils.np_utils import to_categorical
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
Y_train = train[['label']]
X_train = train.drop(["label"],axis=1)
X_test = test
```

```python
X_train.shape
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
Y_train.shape
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.1, random_state=2)
```

```python
clf = MLPClassifier( solver='adam', hidden_layer_sizes=(300,200,300,200,100,10), random_state=1,verbose=True)
#128,64,10
#300,200,300,200,10->143
#300,200,300,200,100,10->98
```

```python

Y_train = to_categorical(Y_train, num_classes = 10)
#Y_val = to_categorical(Y_val, num_classes = 10)
```

```python
print(Y_train.shape)
clf.fit(X_train,Y_train)
```

```python
X_pred_val = clf.predict(X_train)
print("score: ",str(clf.score(X_train,Y_train)))
X_pred_prob = clf.predict_proba(X_train)
X_pred_prob[0:10]
```

```python
Y_test = clf.predict(X_test)
Y_prob = clf.predict_log_proba(X_test)
print(len(Y_test))
print(Y_test.shape)
```

```python
print(Y_test.dtype)
```

```python
l=[]
count=0
index=0
ei=[]
for y,yp in zip(Y_test,Y_prob):
    c=0
    
    for yi in range(10):
        if y[yi] == 1:
            l.append(yi)
            c=1
            break
    if c==0:
        #print(index," ::"+str(np.where(yp==yp.max())))
        l.append(np.where(yp==yp.max())[0][0])
        print(index,"::",l[index])
        count+=1
        ei.append(index)
    index+=1


print(count)
print(index)       
    
```
